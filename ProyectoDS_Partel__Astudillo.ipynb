```python
puntos = df[['release_date', 'publisher', 'name']]
puntos
```

```python
puntos
puntos['release_date'] = pd.to_datetime(puntos['release_date'], errors='coerce')
puntos['year'] = puntos['release_date'].dt.year
conteo_lanzamientos = puntos.groupby(['publisher']).size().reset_index(name='cantidad_lanzamientos')

print(conteo_lanzamientos)
```

```python
top_publishers = conteo_lanzamientos.groupby('publisher')['cantidad_lanzamientos'].sum()

top_publishers = top_publishers.nlargest(5).reset_index()

print(top_publishers)
```

```python
df_topPublishers = pd.DataFrame({
    'publisher': ['Big Fish Games','Strategy First','Ubisoft','THQ Nordic','Square Enix']
})
```

```python
df_filtradoPuntos= puntos[puntos['publisher'].isin(df_topPublishers['publisher'])]

print(df_filtradoPuntos)
```

```python
df_filtradoPuntos.head(50)
```

```python
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(data=df_filtradoPuntos, x = 'year', y = 'publisher')
```

```python
df_counts = df_filtradoPuntos.groupby(['year', 'publisher']).size().reset_index(name='count')
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_counts, x='year', y='count', hue='publisher', marker='o')

plt.xlabel('AÃ±o')
plt.ylabel('NÃºmero de juegos')
plt.title('NÃºmero de juegos lanzados por aÃ±o para los 5 publisher con mas juegos publicados')
plt.legend(title='Publisher', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df_torta = df[['name','achievements']]
df_torta
```

```python
df_torta['achievement_indicator'] = 0
```

```python
df_torta.loc[df_torta['achievements'] > 0, 'achievement_indicator'] = 1
```

```python
df_torta
```

```python
# Contar el nÃºmero de 'achievement_indicator' con valor 1 y 0
achievement_counts = df_torta['achievement_indicator'].value_counts()

# Etiquetas y colores para el grÃ¡fico de torta
labels = ['Con sistema de logros', 'Sin sistema de logros']
colors = ['#66c2a5', '#fc8d62']

# Crear el grÃ¡fico de torta
plt.pie(achievement_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Porcentaje de Usuarios con y sin Sistema de Logros')
plt.show()
```
