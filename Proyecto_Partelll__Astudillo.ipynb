```python
puntos = df[['release_date', 'publisher', 'name']]
puntos
```

```python
puntos
puntos['release_date'] = pd.to_datetime(puntos['release_date'], errors='coerce')
puntos['year'] = puntos['release_date'].dt.year
conteo_lanzamientos = puntos.groupby(['publisher']).size().reset_index(name='cantidad_lanzamientos')

print(conteo_lanzamientos)
```

```python
top_publishers = conteo_lanzamientos.groupby('publisher')['cantidad_lanzamientos'].sum()

top_publishers = top_publishers.nlargest(5).reset_index()

print(top_publishers)
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
#Cargar los datos.
x = df_torta[['achievements']]
y = df_torta[['achievement_indicator']]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

model = DecisionTreeClassifier(random_state=42)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"PrecisiÃ³n del modelo: {accuracy:.2f}")

tree_rules = export_text(model, feature_names=x.columns.tolist())
print("\nReglas del Ã¡rbol de decisiÃ³n:")
print(tree_rules)

plt.figure(figsize=(10, 6))
plot_tree(model, feature_names=x.columns.tolist(), filled=True, rounded=True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Generar datos de ejemplo (2 caracterÃ­sticas para facilitar la visualizaciÃ³n)
X, y = make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0,
                           n_clusters_per_class=1, random_state=42)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear el modelo KNN
k = 5  # NÃºmero de vecinos
knn = KNeighborsClassifier(n_neighbors=k)

# Entrenar el modelo
knn.fit(X_train, y_train)

# Evaluar el modelo
test_accuracy = knn.score(X_test, y_test)
print(f"PrecisiÃ³n en el conjunto de prueba: {test_accuracy:.2f}")

# Visualizar las regiones de decisiÃ³n
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))

# Predecir para cada punto del grid
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Graficar
plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Paired)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.Paired)
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, edgecolor='k', marker='o', label='Test Data')
plt.title(f"K-Nearest Neighbors (k={k})")
plt.xlabel("CaracterÃ­stica 1")
plt.ylabel("CaracterÃ­stica 2")
plt.legend()
plt.show()
```
